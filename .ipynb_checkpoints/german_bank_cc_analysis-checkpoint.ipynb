## Dataset Description:
    The dataset is based on the “Statlog Dataset” from the UCI Machine Learning Repository. 

### Columns of the dataset and their meaning are as follows;

    Age (numeric)
    Sex (text: male, female)
    Job (numeric: 0 - unskilled and non-resident, 1 - unskilled and resident, 2 - skilled, 3 - highly skilled)
    Housing (text: own, rent, or free)
    Saving accounts (text - little, moderate, quite rich, rich)
    Checking account (text - little, moderate, rich)
    Credit amount (numeric, in Deutsche Mark)
    Duration (numeric, in month)
    Purpose (text: car, furniture/equipment, radio/TV, domestic appliances, repairs, education, business, vacation/others

#### 3.1 Outliers

```python
z =np.abs(stats.zscore(df['Age']))
```

```python
threshold = 3
np.where(z>3)
```

```python
threshold = -3
np.where(z < -3)
```

```python
Q1 = np.percentile(df['Age'], 25,interpolation = 'midpoint')
Q2 = np.percentile(df['Age'], 50,interpolation = 'midpoint')
Q3 = np.percentile(df['Age'], 75,interpolation = 'midpoint')
```

```python
IQR = Q3 - Q1
IQR
```

```python
upper = Q3 +1.5*IQR
lower = Q1 - 1.5*IQR
print(lower)
print(upper)
```

```python
outlier = []
for x in df['Age']:
    if (x>upper) or (x<lower):
        outlier.append(x)
```

```python
print(outlier)
```

```python
z1 =np.abs(stats.zscore(df['Credit amount']))

threshold = 3
print(np.where(z1>3))

threshold = -3
print(np.where(z1 < -3))
```

```python
z2 =np.abs(stats.zscore(df['Duration']))

threshold = 3
print(np.where(z2>3))

threshold = -3
print(np.where(z2 < -3))
```

```python
Q11 = np.percentile(df['Credit amount'], 25,interpolation = 'midpoint')
Q21 = np.percentile(df['Credit amount'], 50,interpolation = 'midpoint')
Q31 = np.percentile(df['Credit amount'], 75,interpolation = 'midpoint')

IQR1 = Q3 - Q1
IQR

upper1 = Q31 +1.5*IQR
lower1 = Q11 - 1.5*IQR
print(lower1)
print(upper1)

outlier1 = []
for x in df['Credit amount']:
    if (x>upper1) or (x<lower1):
        outlier1.append(x)

print(outlier1)
```

```python
Q12 = np.percentile(df['Duration'], 25,interpolation = 'midpoint')
Q22 = np.percentile(df['Duration'], 50,interpolation = 'midpoint')
Q32 = np.percentile(df['Duration'], 75,interpolation = 'midpoint')

IQR2 = Q32 - Q12
IQR2

upper2 = Q32 +1.5*IQR
lower2 = Q12 - 1.5*IQR
print(lower2)
print(upper2)

outlier2 = []
for x in df['Duration']:
    if (x>upper2) or (x<lower2):
        outlier2.append(x)

print(outlier2)
```
